```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv
from numpy import linalg
%matplotlib inline
```

```python
mean1 = [0, 0]
cov1 = [[1, 0], [0, 1]]

mean2 = [3, 0]
cov2 = [[1, 0], [0, 1]]

mean3 = [0, 3]
cov3 = [[1, 0], [0, 1]]

n1=int(500*.2)
n2=int(500*.5)
n3=int(500*.3)
```

```python
s1 = np.random.multivariate_normal(mean1, cov1, n1)
s2 = np.random.multivariate_normal(mean2, cov2, n2)
s3 = np.random.multivariate_normal(mean3, cov3, n3)

df = np.concatenate((s1, s2, s3), axis=0)
```

```python
##K=2
k = 2
cents = []
for cluster in range(0,k):
    cents.append(df[np.random.randint(0, len(df), size=1)])
    
loss_list2 = []

for t in range(20):
    ##update clusters
    clusters = [[] for i in range(k)]

    for p in df:  
        mu = min([(i[0], np.linalg.norm(p-cents[i[0]])) for i in enumerate(cents)], key=lambda t:t[1])[0]
        clusters[mu].append(p)

    ##update centroids
    for c in range(k):
        cents[c] = np.mean(clusters[c], axis=0)
    
    ##loss function 
    loss = 0 
    for c in range(k):
        loss += np.linalg.norm(clusters[c]-cents[c])**2
    loss_list2.append(loss)
```

```python
##Question 1
plt.plot(loss_list2, label="$K$=2")
plt.plot(loss_list3, label="$K$=3")
plt.plot(loss_list4, label="$K$=4")
plt.plot(loss_list5, label="$K$=5")
plt.xlabel('Iteration')
plt.ylabel('Loss Function')
plt.title('K-means objective function per iteration')
plt.legend()
```
